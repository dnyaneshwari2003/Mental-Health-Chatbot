TOKANIZATION

```python
import pandas as pd
import os

file_path = "/content/Merged2.csv"
cleaned_file_path = "/content/Merged2_clean.csv"

# Step 1: Check if the file exists
if not os.path.exists(file_path):
    print("Error: File not found!")
else:
    # Step 2: Read and clean file manually
    with open(file_path, 'r', encoding='utf-8', errors='replace') as file:
        lines = file.readlines()

    # Remove bad lines (those with unclosed quotes)
    cleaned_lines = [line for line in lines if line.count('"') % 2 == 0]

    # Step 3: Save the cleaned version
    with open(cleaned_file_path, 'w', encoding='utf-8') as file:
        file.writelines(cleaned_lines)

    print(f"Cleaned file saved as {cleaned_file_path}")

    # Step 4: Try loading the cleaned file with error handling
    try:
        data = pd.read_csv(cleaned_file_path, encoding='utf-8', error_bad_lines=False)
        print("Loaded cleaned CSV successfully!")
        print(data.head())  # Show first 5 rows
    except Exception as e:
        print(f"Error loading cleaned file: {e}")
```

Removing stop words

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words=stopwords.words('english')
```

```python
print(stop_words)
```

Stemming

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
input_str = "This is the input string."  # Assign a value to input_str
input_str=nltk.word_tokenize(input_str)
for word in input_str:
  print(stemmer.stem(word))
```

Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
input_str="This is the input string."
input_str=nltk.word_tokenize(input_str)
for word in input_str:
  print(lemmatizer.lemmatize(word))
```

applying the processign on dataset


```python
leematizer=WordNetLemmatizer()
for index,row in data.iterrows():
  filter_sentence=[]
  sentence=row['Questions']
  sentence=re.sub(r'[^\w\s]','',sentence)
  words=nltk.word_tokenize(sentence)
  words=[w for w in words if not w in stop_words]
  for word in words:
    filter_sentence.append(lemmatizer.lemmatize(word))
  print(filter_sentence)
  # Use .loc instead of .ix for label-based indexing
  data.loc[index,'Questions'] = filter_sentence
```
